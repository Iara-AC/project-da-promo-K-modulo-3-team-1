```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('../datos/hr_raw_data_final.csv', index_col=0)

df.head()
```

```python
#Crear función para redondeo 
columnas_redondeo = ["dailyrate","performancerating"]

def cambio_round (dato):
    return dato.round()

for col in columnas_redondeo:
    df[col] = df[col].apply(cambio_round)
```

```python
# Crear función para transformar string a int

def string_a_int (dato):
    """
    Convierte un número string a int.
    
    Parámetros:
    dato: Número a convertir.
    
    Retorno:
    int: Representación del número en formato int.
    """
    return int(dato)

# Ejemplo de uso
resultado = string_a_int(123.456)
print(resultado)
```
