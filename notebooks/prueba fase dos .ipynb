Las transformaciones se realizarán mediante funciones de Python que se aplicarán a los datos extraídos. Algunas de las transformaciones que podrías hacer son: 

    - La columna `Gender` tiene valores de 0 y 1, los cuales son pocos intutitivos. Los podrías reemplazar por "Male" y "Female", o "M" y "F" por ejemplo. 

    - Algunas columnas, como la columna `DailyRate` son columnas que incluyen valores numéricos decimales. Pero en el DataFrame aparece como columna de tipo string. Deberás hacer los cambios necesarios para convertirla en columna de tipo numérica. 

    - Evaluar si hay valores duplicados y analizar si tiene sentido eliminarlos o mantenerlos. 

    - Valores inconsistentes, por ejemplo en la columna `DistanceFromHome` tiene valores negativos. 

    - Errores tipográficos en algunos valores de las columnas categóricas. Por ejemplo, en la columna `MaritalStatus` en vez de "Married" en algunas filas aparece "Marreid".

```python
#La columna `Gender` tiene valores de 0 y 1, los cuales son pocos intutitivos. Los podrías reemplazar por "Male" y "Female", o "M" y "F" por ejemplo. 
```

```python
# Reemplazar valores en la columna 'Gender'
# Supongamos que 1 representa 'Male' y 0 representa 'Female'
df['gender'] = df['gender'].replace({1: 'Male', 0: 'Female'})


df['gender']
```

```python
# Algunas columnas, como la columna `DailyRate` son columnas que incluyen valores numéricos decimales. Pero en el DataFrame aparece como columna de tipo string. Deberás hacer los cambios necesarios para convertirla en columna de tipo numérica.
```

```python
# Convertir la columna 'DailyRate' de string a tipo numérico (float)
df['dailyrate'] = pd.to_numeric(df['dailyrate'], errors='coerce')
df['dailyrate']
```

```python
#Evaluar si hay valores duplicados y analizar si tiene sentido eliminarlos o mantenerlos. 
#Valores inconsistentes, por ejemplo en la columna `DistanceFromHome` tiene valores negativos.

# Verifico duplicados
duplicados = df.duplicated().sum()
print(f"Número de registros duplicados: {duplicados}")
```

```python

#Errores tipográficos en algunos valores de las columnas categóricas. Por ejemplo, en la columna `MaritalStatus` en vez de "Married" en algunas filas aparece "Marreid". 

# Valores únicos en MaritalStatus
print("Valores únicos en 'maritalstatus':", df['maritalstatus'].unique())

```

```python
# Reemplazar errores tipográficos
df['maritalstatus'] = df['maritalstatus'].replace({
    "marreid": "married",
    "singel": "single",
    "divorcd": "divorced"
    # Añade más reemplazos según sea necesario
})
df['maritalstatus']
```

```python
# Crear un diccionario de mapeo para correcciones
#Define un diccionario donde las claves son los valores correctos y los valores son listas de los errores tipográficos comunes asociados. 
#Luego, el diccionario se invertirá para usar en la corrección
correcciones = {
    "Married": ["Marreid", "Marrid", "Mairred"],
    "Single": ["Singel", "Sigle"],
    "Divorced": ["Divorcd", "Divrced"],
    "Widowed": ["Widwed", "Widoved"]
}
```

```python
# Para facilitar el reemplazo, invertimos el diccionario para que cada error tipográfico apunte al valor correcto.
correcciones_invertido = {error: correcto for correcto, errores in correcciones.items() for error in errores}

correcciones_invertido
```

```python
# Reemplazar errores tipográficos en la columna 'MaritalStatus'
df['maritalstatus'] = df['maritalstatus'].replace(correcciones_invertido)
df['maritalstatus']
```

```python
#Columnas redundantes, es decir, columnas que nos dan la misma información expresada de forma diferente.

# Eliminar columnas duplicadas
df = df.loc[:, ~df.T.duplicated()]  #

# Verificar resultado
print("Columnas restantes después de eliminar duplicadas:", df.columns)
```
